## Re-Shuffling the dataset:

## Load necessary libraries

```r
library(ggplot2)
library(dplyr)
library(tidyr)
library(viridis)
```

## Define age bins and labels

```r
age_bins <- c(0, 20, 40, 60, 80, 100)
age_labels <- c('0-20', '21-40', '41-60', '61-80', '81-100')
```

```r
shuffled_df$AgeGroup <- cut(shuffled_df$age, breaks = age_bins, labels = age_labels, right = FALSE)
```

## Define BMI bins and labels

```r
bmi_bins <- c(10, 20, 30, 40, 50, 60, 70, 80, 100)
bmi_labels <- c('10-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-100')
```

```r
shuffled_df$BMICategory <- cut(shuffled_df$bmi, breaks = bmi_bins, labels = bmi_labels, right = FALSE)
```

## Define glucose level bins and labels

```r
glucose_bins <- c(50, 100, 150, 200, 250, 300)  
glucose_labels <- c('50-100', '101-150', '151-200', '201-250', '251-300')  
```

```r
shuffled_df$GlucoseRange <- cut(shuffled_df$avg_glucose_level, breaks = glucose_bins, labels = glucose_labels, right = FALSE)
```

## Calculate stroke rate for each combination of AgeGroup and BMICategory

```r
stroke_rate <- shuffled_df %>%
  group_by(AgeGroup, BMICategory) %>%
  summarise(stroke_rate = mean(stroke) * 100, .groups = "drop")  # Use .groups = "drop" to avoid keeping groups


stroke_rate_pivot <- stroke_rate %>%
  spread(key = BMICategory, value = stroke_rate)

# Creating the plot for AgeGroup and BMICategory:
ggplot(stroke_rate, aes(x = BMICategory, y = AgeGroup, fill = stroke_rate)) +
  geom_tile() +
  geom_text(aes(label = sprintf("%.1f", stroke_rate)), color = "white", size = 4) +
  scale_fill_gradient(low = "lightblue", high = "darkblue") +
  labs(title = "Stroke Rate by Age Group and BMI Category",
       x = "BMI Category",
       y = "Age Group",
       fill = "Stroke Rate (%)") +
  theme_minimal()
```

## Calculate stroke percentage for each combination of AgeGroup and GlucoseRange

```r
stroke_percentage <- shuffled_df %>%
  group_by(AgeGroup, GlucoseRange) %>%
  summarise(stroke_rate = mean(stroke) * 100, .groups = "drop")


heatmap_data <- stroke_percentage %>%
  spread(key = GlucoseRange, value = stroke_rate)

# Creating the plot for AgeGroup and GlucoseRange:
ggplot(stroke_percentage, aes(x = GlucoseRange, y = AgeGroup, fill = stroke_rate)) +
  geom_tile() +
  geom_text(aes(label = sprintf("%.2f", stroke_rate)), color = "white", size = 4) +
  scale_fill_gradient(low = "lightblue", high = "darkblue") +  # Color palette
  labs(title = "Stroke Percentage by Age Group and Glucose Level",
       x = "Glucose Level Group",
       y = "Age Group",
       fill = "Stroke Percentage (%)") +
  theme_minimal()

```

# Calculate point-biserial correlation (using Pearson correlation)

```r
# First, we split the dataframe into binary and numerical columns
binary_columns <- shuffled_df[, 4:ncol(df)]  
numerical_columns <- shuffled_df[, 1:3]      

# Now, we are create an empty matrix to store the resulting correlations
correlation_matrix <- matrix(NA, nrow = ncol(binary_columns), ncol = ncol(numerical_columns))
rownames(correlation_matrix) <- colnames(binary_columns)
colnames(correlation_matrix) <- colnames(numerical_columns)

# Now, we calculate the correlations for each of the binary columns with each of the numerical column using Point-Biserial correlation
for (i in 1:ncol(binary_columns)) {
  for (j in 1:ncol(numerical_columns)) {
    
    result <- cor.test(binary_columns[, i], numerical_columns[, j], method = "pearson")
    correlation_matrix[i, j] <- result$estimate  
  }
}

# Converting the correlation matrix to a data frame
correlation_df <- as.data.frame(correlation_matrix)


print(correlation_df)

```

# Calculate the correlation between stroke and each other binary column

```r
correlation_with_stroke <- sapply(binary_columns, function(x) cor(x, binary_columns$stroke, method = "pearson"))

print(correlation_with_stroke)
```
